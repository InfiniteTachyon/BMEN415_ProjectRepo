```python
import numpy as np
import pandas as pd  # Importing Pandas library

# read the excel file
data = pd.read_excel('Volumetric_features.xlsx')

# convert to csv format
data.to_csv('Volumetric_features.csv', index=False)

data #displaying the data
```

```python
data.info() #getting information about data
```

```python
data.describe()
```

```python
data = data.dropna()
x = data.drop(["Age", "S.No", "dataset"], axis=1)
y = data.Age.values
```

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Create the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

accuracy_train = r2_score(y_train, y_pred_train)
print("Training Accuracy for Linear Regression Model: ", accuracy_train)

accuracy_test = r2_score(y_test, y_pred_test)
print("Testing Accuracy for Linear Regression Model: ", accuracy_test)

RMSE_train = sqrt(mean_squared_error(y_train, y_pred_train))
print("RMSE for Training Data: ", RMSE_train)

RMSE_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("RMSE for Testing Data: ", RMSE_test)

rmse_norm = np.sqrt(mean_squared_error(y_test,y_pred_test))/(max(y_test)-min(y_test))
print(rmse_norm)
```

```python
import matplotlib.pyplot as plt 

true_value = y_train
predicted_value = y_pred_train

plt.figure(figsize=(10,10))
plt.scatter(true_value, predicted_value, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(predicted_value), max(true_value))
p2 = min(min(predicted_value), min(true_value))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predicted', fontsize=15)
plt.title("True Values vs Predicted ")
plt.axis('equal')
plt.show()
```
